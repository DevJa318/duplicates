```python
import hashlib

def hashfile(file: str, block_size: int = 65536) -> str:
    """Generate the hash of any file according to the sha256 algorithm."""
    with open(file, 'rb') as message:
        m = hashlib.sha256()
        block = message.read(block_size)
        while len(block) > 0:
            m.update(block)
            block = message.read(block_size)
        digest = m.hexdigest()

    return digest
```

```python
%%time
def list_hashes_and_files(files_register: str) -> list: 
    """Generate sorted list of tuples with hashes and files' adresses"""
    register_of_hashes = []
    with open(files_register,'r') as f:
        for adres in f:
            files_sha = hashfile(adres.strip('\n'))
            register_of_hashes.append((files_sha, adres))
    return sorted(register_of_hashes)
```
